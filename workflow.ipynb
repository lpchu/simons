## 1. Read, parse & clean input data
- Merge all files under `exercise_input_data_public/` into 1 single table.
- The table should be made up of 9 columns x samples.
- 9 variant info columns: `#CHROM`, `POS`, `ID`, `REF`, `ALT`, `QUAL`, `FILTER`, `INFO` and `FORMAT`.
- Update sample "HG10101" to be "HG00101".

```python
# 2. Merge all files
with open('all_files_names.txt', 'r') as f:
    file_names = [l.strip() for l in f]

df_final = pd.DataFrame()
for name in tqdm(file_names):
    df = pd.read_csv(path+name, sep='\t', compression='gzip', dtype=object)
    df_final = pd.concat([df_final, df], ignore_index=True)
```

```python
# 3. Update sample "HG10101" to be "HG00101"
df_final.rename(columns={"HG10101": "HG00101"}, inplace=True)
df_final.columns
```

## 2. Count alleles
Create a new table with major & minor allele counts per sample, with columns: (1) sample_id, (2) major_count, & (3) minor_count, and save as “allele_counts.tsv” (TSV file with no index column). 0=major allele & 1=minor allele (genotype “1|1” = 2 minor alleles).

```python
sample_ids = df_final.columns[9:]
len(sample_ids)
```

```python
allele_counts = pd.DataFrame(columns=['sample_id', 'major_count', 'minor_count'])
allele_counts['sample_id'] = sample_ids
allele_counts = allele_counts.set_index('sample_id')
for sample_id in tqdm(sample_ids):
    allele_counts.loc[sample_id, 'major_count'] = df_final[sample_id].apply(lambda x: x.split('|').count('0') if not pd.isnull(x) else np.nan).sum()
    allele_counts.loc[sample_id, 'minor_count'] = df_final[sample_id].apply(lambda x: x.split('|').count('1') if not pd.isnull(x) else np.nan).sum()
```

```python
allele_counts.head()
```

```python
# save file
allele_counts.to_csv('allele_counts.tsv', sep='\t', index=False)
```

## 3. Subnet and save new files
From the table / data structure you created above, save the data as 10 different files based on the last
digit of the sample IDs (e.g. all samples named HG*3 would be included in “output/3.chr21.10000000_14999999.tsv.gz”). Also include all variant info columns in each file.

```python
!mkdir -p output
```

```python
df_final.head()
```

```python
variant_info_columns = ['#CHROM', 'POS', 'ID', 'REF', 'ALT', 'QUAL', 'FILTER', 'INFO', 'FORMAT']
for i in tqdm(range(10)):
    selected_samples = [x for x in sample_ids if x.endswith(str(i))]
    selected_cols = variant_info_columns + selected_samples
    df = df_final[selected_cols].copy()
    df.dropna(how='all', subset=selected_samples, inplace=True)
    output_file = 'output/' + f"{i}.chr21.10000000_14999999.tsv.gz"
    df.to_csv(output_file, sep='\t', compression='gzip', index=False)
```
